**Reusable Cleaning functions**

**Function 1 - Clean Column Names**

```python
def clean_column_names(df):

    df.columns=(
        df.columns.astype(str)
        .str.strip()
        .str.lower()
        .str.replace(" ","_")
        .str.replace("-","_")
    )

    return df
```

**Function 2 - Clean Year Columns**

```python
def clean_year_columns(df):

    df.rename(columns={
        '2022.0':'2022',
        '2023.0':'2023',
        '2024.0':'2024'},inplace =True) 

    return df
```

**Function 3 Remove Total Rows**

```python
def remove_total_rows(df):

    df=df[
        ~df['city'].astype(str).str.contains('total',case=False, na=False)]

    return df
```

**Function 4 Drop Empty Rows**

```python
def remove_empty_rows(df):

    df.dropna(how='all', inplace=True)

    return df
```

**Function 5 Drop Unnecessary Columns**

```python
def remove_unwanted_columns(df):
    if 'sl' in df.columns:
        df = df.drop(columns=['sl'])
        
    return df
```

**Function 6 Melt Data**

```python
def melt_crime_data(df, category_name, rate_column):

    df_long = df.melt(
        id_vars=[
            'city',
            'actual_population_(lakhs)_2011',
            rate_column,
            'chargesheeting_rate_(2024)'
        ],
        value_vars=['2022', '2023', '2024'],
        var_name='year',
        value_name='cases'
    )

    # Standardize column names
    df_long.rename(columns={
        'actual_population_(lakhs)_2011': 'population_lakhs_2011',
        rate_column: 'crime_rate_2024',
        'chargesheeting_rate_(2024)': 'chargesheeting_rate_2024'
    }, inplace=True)

    df_long['category'] = category_name

    return df_long
```

**Function 7 - Final Cleaning**

```python


def final_cleaning(df):
   
    df = df.dropna()

    df = df.drop_duplicates()
  
    df = df.reset_index(drop=True)

    df['year'] = df['year'].astype(int)
    
    return df
```

**Function 8 - Extract City State**

```python
def extract_city_state(df):

    # Create state column
    df['state'] = df['city'].str.extract(r'\((.*?)\)')

    # Clean city column
    df['city'] = df['city'].str.replace(
        r'\s*\(.*?\)',
        '',
        regex=True
    )

    # Handle Delhi separately
    df.loc[
        df['city'] == 'Delhi City',
        'state'
    ] = 'Delhi'

    return df
```

**Reorder columns**

```python
def reorder_columns(df):
    df = df[
    [
        'year',
        'state',
        'city',
        'category',
        'cases',
        'population_lakhs_2011',
        'crime_rate_2024',
        'chargesheeting_rate_2024'
    ]]
    return df
```
